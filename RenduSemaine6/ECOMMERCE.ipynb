```python
import pandas as pd
from functools import reduce
# Supposons que vous avez les DataFrames commandes, clients, produits, lignescommande, et categories
dataframes = [commandes, clients, produits, lignescommande, categories]

# Utiliser reduce avec la fonction concat pour fusionner les DataFrames
Ecommerce = reduce(lambda left, right: pd.concat([left, right], axis=1), dataframes)

# Affichage des premières lignes du DataFrame résultant
print(Ecommerce.head())
```

## les 20 produits les plus vendue 

###   SELECT p.nomProduit, SUM(lc.quantite) AS TotalQuantiteVendue
###   FROM Produit p
###   JOIN LignesCommande lc ON p.idProduit = lc.idProduit
###   GROUP BY p.nomProduit
###   ORDER BY TotalQuantiteVendue DESC
###   LIMIT 20;

#### CODE PYTHON

```python
# Groupement par produit et calcul de la somme des quantités vendues
top_products = Ecommerce.groupby('nomProduit')['quantite'].sum().reset_index()

# Tri par quantité vendue de manière décroissante
top_products = top_products.sort_values(by='quantite', ascending=False)

# Sélection des 20 premiers produits
top_20_products = top_products.head(20)

# Affichage des résultats
print(top_20_products)
```

#### Visualisation avec seaborn

```python

plt.figure(figsize=(12, 8))
sns.barplot(x='quantite', y='nomProduit', data=top_20_products, palette='viridis')
plt.title('Les 20 produits les plus vendus')
plt.xlabel('Quantité vendue')
plt.ylabel('Produit')
plt.show()
```

# Répartition des montants dépensés par mois

## SELECT MONTH(dateCommande) AS MoisCommande, SUM(prix) AS MontantTotalVentes
## FROM Ecommerce
## GROUP BY MoisCommande;


```python
# Convertir la colonne 'dateCommande' en format de date
Ecommerce['dateCommande'] = pd.to_datetime(Ecommerce['dateCommande'])

# Extraire le mois de la colonne 'dateCommande'
Ecommerce['MoisCommande'] = Ecommerce['dateCommande'].dt.month

# Calcul du montant total des ventes par mois
ventes_par_mois = Ecommerce.groupby('MoisCommande')['prix'].sum().reset_index()

# Visualisation avec un barplot
plt.figure(figsize=(12, 8))
plt.bar(ventes_par_mois['MoisCommande'], ventes_par_mois['prix'], color='purple')
plt.title('Répartition des montants dépensés par mois')
plt.xlabel('Mois de commande')
plt.ylabel('Montant total des ventes')
plt.show()
```

# le nombre de commandes par statut

## SELECT statut, COUNT(DISTINCT idCommande) AS NombreCommandes
## FROM Ecommerce
## GROUP BY statut;

```python
# Calcul du nombre de commandes par statut
commandes_par_statut = Ecommerce.groupby('statut')['idCommande'].nunique().reset_index()

# Visualisation avec un barplot
plt.figure(figsize=(12, 8))
plt.bar(commandes_par_statut['statut'], commandes_par_statut['idCommande'], color='blue')
plt.title('Nombre de commandes par statut')
plt.xlabel('Statut de commande')
plt.ylabel('Nombre de commandes')
plt.show()
```
